````python
# usando la conn que ya existe:

with conn.cursor() as cursor:
    cursor.execute('''
    SELECT customerName, amount
    FROM customers inner join payments
    ON payments.customerNumber = customers.customerNumber
    ORDER BY amount DESC
    LIMIT 10
    ''')
    dbs = cursor.fetchall()
dbs
````

3. Escribe una consulta que, para cada producto comprado muestre la clave del cliente `customerNumber` y el código del producto `productCode`. Obten los primeros 10 valores

````python
#Buscar nombres de tablas
with conn.cursor() as cursor:
    cursor.execute('''
    SELECT TABLE_NAME, COLUMN_NAME
    FROM INFORMATION_SCHEMA.COLUMNS 
    WHERE 
        TABLE_SCHEMA = 'nabla_python1'
    ''')
    dbs = cursor.fetchall()
dbs
````

````python
#obtener foreign keys y sus asociaciones a primary keys
with conn.cursor() as cursor:
    cursor.execute('''
SELECT
    `column_name`, 
    `referenced_table_schema` AS foreign_db, 
    `referenced_table_name` AS foreign_table, 
    `referenced_column_name`  AS foreign_column 
FROM
    `information_schema`.`KEY_COLUMN_USAGE`
WHERE
    `constraint_schema` = SCHEMA()
AND
    `table_name` = 'orders'
AND
    `referenced_column_name` IS NOT NULL
ORDER BY
    `column_name`;
    ''')
    dbs = cursor.fetchall()
dbs
````

````python
# usando la conn que ya existe:

with conn.cursor() as cursor:
    cursor.execute('''
    SELECT customers.customerNumber, products.productCode
    FROM products inner join orderdetails
    ON orderdetails.productCode = products.productCode inner join orders
    ON orders.orderNumber = orderdetails.orderNumber inner join customers
    ON customers.customerNumber = orders.customerNumber
    LIMIT 10
    ''')
    dbs = cursor.fetchall()
dbs
````

6. Con base en ejercicio 3, encuentra el nombre del producto `productName` y cuantas unidades se han comprado de ese producto. Ordena los elementos de mayor a menor.

````python

with conn.cursor() as cursor:
    cursor.execute('''
    SELECT products.productName, orderdetails.quantityOrdered
    FROM products inner join orderdetails
    ON orderdetails.productCode = products.productCode inner join orders
    ON orders.orderNumber = orderdetails.orderNumber inner join customers
    ON customers.customerNumber = orders.customerNumber
    ORDER BY orderdetails.quantityOrdered DESC
    ''')
    dbs = cursor.fetchall()
dbs
````
