#### 5. Convierta la matriz de covarianza en una matriz de correlaci√≥n usando Python

$$\sigma_{x,y}=cov(x,y)=\frac{\sum_{i=1}^{n}(x_i-\bar{x})\sum_{i=1}^{n}(y_i-\bar{y})}{n}$$

$$ corr(x,y)=\frac{\sigma_{x,y}}{\sigma_x\cdot \sigma_y}$$

```python
!git clone https://github.com/pikey-msc/Python.git
```

```python
import pandas as pd
dataset = pd.read_csv("Python/M3/S2/iris.csv") 
  
dataset.head() 
```

Quitamos la columna destino

```python
data = dataset.drop(["Id","Species"], axis='columns' ).values
#data=dataset.iloc[:, :-5]
data
data.shape
```

```python
#programa la matriz de var-cov
def calcCov(x, y): 
  
  mean_x, mean_y = x.mean(), y.mean() 
  n = len(x) 

  return sum((x - mean_x) * (y - mean_y)) / (n -1)
  
  
def covMat(data): 
    
  rows, cols = data.shape 

  
  
  cov_mat = np.zeros((cols, cols)) 

  for i in range(cols): 

      for j in range(cols): 
          
          cov_mat[i][j] = calcCov(data[:, i], data[:, j]) 

  return cov_mat 
  
covMat(data) 
```

```python
#comprobando con np.cov
np.cov(data,rowvar=False)
```

```python
#Calculando correlaci√≥n
def corrMat(data): 
  
  rows, cols = data.shape 

  corr_mat = np.zeros((cols, cols)) 

  for i in range(cols): 

      for j in range(cols): 

          x, y = data[:, i], data[:, j] 
          
          corr_mat[i][j] = calcCov(x, y) / (x.std() * y.std()) 

  return corr_mat 
corrMat(data)
```

```python
# define la funci√≥n a integrar
def f(x):
    return x
```

```python
x_lower = 0 # the lower limit of x
x_upper = 1 # the upper limit of x

val, abserr = quad(f, x_lower, x_upper)

print( "integral value =", val, ", absolute error =", abserr )
```

```python
def integrand(x, n):
    """
    Bessel function of first kind and order n. 
    """
    return jn(n, x)


x_lower = 0  # the lower limit of x
x_upper = 10 # the upper limit of x

val, abserr = quad(integrand, x_lower, x_upper, args=(3,))

print (val, abserr )
```

## Ecuaciones diferenciales ordinarias (ODEs)

SciPy provee dos maneras diferentes de resolver ODEs (Ecuaciones diferenciales ordinarias): Una basada en API llamada `odeint`, y una API orientada a objetos sobre la clase `ode`. Usualmente la `odeint` es m√°s facil de usar pero la de `ode` ofrece mejores niveles de contro.




```python
from scipy.integrate import odeint, ode
```

Un sistema de ODE son formulados de manera est√°ndar como sigue:

$y' = f(y, t)$

donde

$y = [y_1(t), y_2(t), ..., y_n(t)]$ 

y $f$ es la funci√≥n que da las derivadas de $y_i(t)$. Para resolver una ODE necesitamos la forma $f$ y la condici√≥n inicial, $y(0)$.

Una vez que hayamos definido la funci√≥n `f` y el array `y_0` (esto es $f$ y $y(0)$ ), usamos `odeint` como:

    y_t = odeint(f, y_0, t)

donde `t` es un array con coordenadas de tiempo para resolver el problema de ODE. `y_t` es un arreglo con una fila para cada punto en el tiempo `t`, donde cada columna corresponde a una soluci√≥n `y_i(t)` al punto en el tiempo



#### Ejemplo: Oscilador arm√≥nico amortiguado

Este problema se puede ver a detalle en http://en.wikipedia.org/wiki/Damping

La ecuaci√≥n del movimiento del oscilador amortiguado es:


$\displaystyle \frac{\mathrm{d}^2x}{\mathrm{d}t^2} + 2\zeta\omega_0\frac{\mathrm{d}x}{\mathrm{d}t} + \omega^2_0 x = 0$

donde $x$ es la posici√≥n del oscilador, $\omega_0$ is la frecuencia, y $\zeta$ es el ratio de amortiguaci√≥n. Para escribir una ODE de segundo orden de forma estandar intriducimos $p = \frac{\mathrm{d}x}{\mathrm{d}t}$:

$\displaystyle \frac{\mathrm{d}p}{\mathrm{d}t} = - 2\zeta\omega_0 p - \omega^2_0 x$

$\displaystyle \frac{\mathrm{d}x}{\mathrm{d}t} = p$

Introduciremos argumentos a la funci√≥n  `odeint`:

```python
def dy(y, t, zeta, w0):
    """
    The right-hand side of the damped oscillator ODE
    """
    x, p = y[0], y[1]
    
    dx = p
    dp = -2 * zeta * w0 * p - w0**2 * x

    return [dx, dp]
```

```python
# initial state: 
y0 = [1.0, 0.0]
```

```python
# time coodinate to solve the ODE for
t = linspace(0, 10, 1000)
w0 = 2*pi*1.0
```

```python
# solve the ODE problem for three different values of the damping ratio

y1 = odeint(dy, y0, t, args=(0.0, w0)) # undamped
y2 = odeint(dy, y0, t, args=(0.2, w0)) # under damped
y3 = odeint(dy, y0, t, args=(1.0, w0)) # critial damping
y4 = odeint(dy, y0, t, args=(5.0, w0)) # over damped
```

```python
fig, ax = plt.subplots()
ax.plot(t, y1[:,0], 'k', label="undamped", linewidth=0.25)
ax.plot(t, y2[:,0], 'r', label="under damped")
ax.plot(t, y3[:,0], 'b', label=r"critical damping")
ax.plot(t, y4[:,0], 'g', label="over damped")
ax.legend();
```

### Eigenvalores y eigenvectores


Los eigenvalores de una matriz $A$ provienen de :

$\displaystyle A v_n = \lambda_n v_n$

donde $v_n$ es el vector $n$-√©simo eigenvector y $\lambda_n$ es el $n$-√©simo eigenvalue.

Para solo eigenvalores usar `eigvals` y para ambos usar `eig`:

```python
evals = eigvals(A)
```

```python
evals
```

```python
evals, evecs = eig(A)
```

```python
evals
```

```python
evecs
```


Los eigenvectores corresponden al $n$√©simo eigenvalor (guardado en `evals[n]` es la columna $n$ en `evecs`, i.e., `evecs[:,n]`. Para verificar esto, multiplicaremos los eigenvectores con la matriz y comparamos el producto del eigenvector y el eigenvalor. 

```python
n = 2

norm(dot(A, evecs[:,n]) - evals[n] * evecs[:,n])
```

## Interpolaci√≥n

Interpolaci√≥n es simple y conveniente: La funci√≥n `interp1d`, cuando recibe arreglos con X y y, retorna la interpolaci√≥n.

```python
from scipy.interpolate import *
```

```python
def f(x):
    return sin(x)
```

```python
n = arange(0, 10)  
x = linspace(0, 9, 100)

y_meas = f(n) + 0.1 * randn(len(n)) # simulate measurement with noise
y_real = f(x)

linear_interpolation = interp1d(n, y_meas)
y_interp1 = linear_interpolation(x)

cubic_interpolation = interp1d(n, y_meas, kind='cubic')
y_interp2 = cubic_interpolation(x)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(n, y_meas, 'bs', label='noisy data')
ax.plot(x, y_real, 'k', lw=2, label='true function')
ax.plot(x, y_interp1, 'r', label='linear interp')
ax.plot(x, y_interp2, 'g', label='cubic interp')
ax.legend(loc=3);
```

## Estad√≠stica

El m√≥dulo `scipy.stats` contiene un gran n√∫mero de distribuciones, funciones estad√≠sticas y pruebas. Ver http://docs.scipy.org/doc/scipy/reference/stats.html.

Existe tambi√©n un paquete llamado statsmodels. Ver http://statsmodels.sourceforge.net 

```python
from scipy import stats
```

```python
# create a (discrete) random variable with Poissonian distribution

X = stats.poisson(3.5) # photon distribution for a coherent state with n=3.5 photons
```

```python
n = arange(0,15)

fig, axes = plt.subplots(3,1, sharex=True)

# plot the probability mass function (PMF)
axes[0].step(n, X.pmf(n))

# plot the cumulative distribution function (CDF)
axes[1].step(n, X.cdf(n))

# plot histogram of 1000 random realizations of the stochastic variable X
axes[2].hist(X.rvs(size=1000));
```

```python
# create a (continous) random variable with normal distribution
Y = stats.norm()
```

```python
x = linspace(-5,5,100)

fig, axes = plt.subplots(3,1, sharex=True)

# plot the probability distribution function (PDF)
axes[0].plot(x, Y.pdf(x))

# plot the cumulative distribution function (CDF)
axes[1].plot(x, Y.cdf(x));

# plot histogram of 1000 random realizations of the stochastic variable Y
axes[2].hist(Y.rvs(size=1000), bins=50);
```

Estad√≠sticas:

```python
X.mean(), X.std(), X.var() # Poisson distribution
```

```python
Y.mean(), Y.std(), Y.var() # normal distribution
```

### Pruebas estad√≠sticas

Probar si dos conjuntos vienen de la misma distribuci√≥n:

```python
t_statistic, p_value = stats.ttest_ind(X.rvs(size=1000), X.rvs(size=1000))

print( "t-statistic =", t_statistic)
print( "p-value =", p_value)
```

Dado que el p-value es muy grande no se puede rechazar la hip√≥tesis d eque dos conjuntos de valores aleatorios tengan diferente media.


Para probar si la media de una muestra de datos tiene media 0.1, la media verdadera es 0


```python
stats.ttest_1samp(Y.rvs(size=1000), 0.1)
```

Se puede rechazar la hip√≥tesis, p-value bajo.


```python
Y.mean()
```

```python
stats.ttest_1samp(Y.rvs(size=1000), Y.mean())
```
